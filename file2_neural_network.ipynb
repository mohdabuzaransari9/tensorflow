**improving the model**

in the steps **creating the model**
1. you can add more layers
2. add more hidden units in layers
3. in each layer change activation function

**in compiling the model**
1. change optimizer----> it can be SGD() or Adam
2. In Adam you can also change learning rate change learning <br>(how much model should improve in each step)

**in fitting the model**
1. you can change the epochs value to learn from more examples

```python
#createing model
import tensorflow as tf
tf.random.set_seed(42)
model=tf.keras.Sequential([   #Sequential m S capital
    tf.keras.layers.Dense(1)

])

# compiling the model
model.compile(
    loss=tf.keras.losses.mae,
    optimizer=tf.keras.optimizers.SGD(),
    metrics=["mae"]
)

# fitting the model
model.fit(tf.expand_dims(x,axis=-1),y,epochs=100)
```

```python
model.predict([np.array([50])])
```

**as our result should be 60 but predicted result its far from the actual result**

```python
# but its getting better than previous one
```

**improving much more**

```python
# creating model
model=tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation='relu'),
    tf.keras.layers.Dense(1)

])

# compile the model
model.compile(
    loss="mae",# can also be wriiten as
    optimizer=tf.keras.optimizers.SGD(),
    metrics=["mae"]# it should be in list
)
#fitting the model
model.fit(tf.expand_dims(x,axis=-1),y,epochs=100)
```

visualizing on scater plot

```python
def plot_visualization(training_data=x_train,
                   testing_data=x_test,
                   training_label=y_train,
                   testing_label=y_test,
                   predicted_data=y_pred):
  '''plotting scatter plot using training_data , testing_data, training_label, testing_label,predicted_data'''

  import matplotlib.pyplot as plt
  plt.figure(figsize=(10,7))
  plt.scatter(training_data,training_label,c="b",label="training data")
  plt.scatter(testing_data,testing_label,c="g",label="testing data")
  plt.scatter(testing_data,predicted_data,c="r",label="predicted_data")
  plt.legend()
  plt.show()


```

```python
plot_visualization(training_data=x_train,
                   testing_data=x_test,
                   training_label=y_train,
                   testing_label=y_test,
                   predicted_data=y_pred)
```

**creating function for mae**

```python
def mae(y_true,y_pred):
  mae=tf.metrics.mae(y_true=y_test,y_pred=tf.squeeze(y_pred))
  return mae
```

**creating function for mse**

```python
def mse(y_true,y_pred):
  mse=tf.metrics.mse(y_true,y_pred=tf.squeeze(y_pred))
  return mse
```
