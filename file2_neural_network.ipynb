```python
import numpy as np
# features(independent variables)
x=np.array([-7.0,-4.0,-1.0,2.0,5.0,8.0,11.0,14.0])
# outcome(dependent variables)
y=np.array([3.,6.,9.,12.,15.,18.,21.,24.])
```

```python
#convertingthese into tensors
import tensorflow as tf
x=tf.constant(tf.cast(x,dtype=tf.float32))
y=tf.constant(tf.cast(y,dtype=tf.float32))
x,y
```

**what is loss?**<br>

ytrue-ypred ka absolute andthen is ka mean<br>

loss=mean(abs(ytue-ypred))

```python
#createing model
import tensorflow as tf
tf.random.set_seed(42)
model=tf.keras.Sequential([   #Sequential m S capital
    tf.keras.layers.Dense(1)

])
```

```python
# compiling the model
model.compile(
    loss=tf.keras.losses.mae,
    optimizer=tf.keras.optimizers.SGD(),
    metrics=["mae"]
)
```

```python
# fitting the model
model.fit(tf.expand_dims(x,axis=-1),y,epochs=5)
```

```python
model.predict([np.array([10])])
```

```python
model.predict([np.array([50])])
```

**improving the model**

in the steps **creating the model**
1. you can add more layers
2. add more hidden units in layers
3. in each layer change activation function

**in compiling the model**
1. change optimizer----> it can be SGD() or Adam
2. In Adam you can also change learning rate change learning <br>(how much model should improve in each step)

**in fitting the model**
1. you can change the epochs value to learn from more examples

```python
#createing model
import tensorflow as tf
tf.random.set_seed(42)
model=tf.keras.Sequential([   #Sequential m S capital
    tf.keras.layers.Dense(1)

])

# compiling the model
model.compile(
    loss=tf.keras.losses.mae,
    optimizer=tf.keras.optimizers.SGD(),
    metrics=["mae"]
)

# fitting the model
model.fit(tf.expand_dims(x,axis=-1),y,epochs=100)
```

```python
model.predict([np.array([50])])
```

**as our result should be 60 but predicted result its far from the actual result**

```python
# but its getting better than previous one
```

**improving much more**

```python
# creating model
model=tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation='relu'),
    tf.keras.layers.Dense(1)

])

# compile the model
model.compile(
    loss="mae",# can also be wriiten as
    optimizer=tf.keras.optimizers.SGD(),
    metrics=["mae"]# it should be in list
)
#fitting the model
model.fit(tf.expand_dims(x,axis=-1),y,epochs=100)
```

```python
model.predict([np.array([50])])
```

**losses and mae is is reduced to nearly 3**
