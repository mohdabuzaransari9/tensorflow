2. In Adam you can also change learning rate change learning <br>(how much model should improve in each step)

**in fitting the model**
1. you can change the epochs value to learn from more examples

```python
#createing model
import tensorflow as tf
tf.random.set_seed(42)
model=tf.keras.Sequential([   #Sequential m S capital
    tf.keras.layers.Dense(1)

])

# compiling the model
model.compile(
    loss=tf.keras.losses.mae,
    optimizer=tf.keras.optimizers.SGD(),
    metrics=["mae"]
)

# fitting the model
model.fit(tf.expand_dims(x,axis=-1),y,epochs=100)
```

visualizing on scater plot

```python
def plot_visualization(training_data=x_train,
                   testing_data=x_test,
                   training_label=y_train,
                   testing_label=y_test,
                   predicted_data=y_pred):
  '''plotting scatter plot using training_data , testing_data, training_label, testing_label,predicted_data'''

  import matplotlib.pyplot as plt
  plt.figure(figsize=(10,7))
  plt.scatter(training_data,training_label,c="b",label="training data")
  plt.scatter(testing_data,testing_label,c="g",label="testing data")
  plt.scatter(testing_data,predicted_data,c="r",label="predicted_data")
  plt.legend()
  plt.show()


```

```python
insurance=pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/refs/heads/master/insurance.csv")
insurance
```

```python
from sklearn.compose import make_column_transformer
```

```python
from sklearn.preprocessing import MinMaxScaler

from sklearn.preprocessing import OneHotEncoder
```

```python
ct=make_column_transformer(
                    (MinMaxScaler(),["age",'bmi',"children"]),
                    (OneHotEncoder(handle_unknown="ignore"),["sex","smoker","region"])

)
```

```python
# # create x and y from start
x=insurance.drop("charges",axis=1)
y=insurance["charges"]
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
ct.fit(x_train)
```

```python
x_train_normal=ct.transform(x_train)
x_test_normal=ct.transform(x_test)
```

ct.jpg

```python
x_train_normal[0]
```

```python
tf.random.set_seed(42)
insurance_model3=tf.keras.Sequential([

  tf.keras.layers.Dense(10),
  tf.keras.layers.Dense(1)
])


insurance_model3.compile(
    loss=tf.keras.losses.mae,
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
    metrics=["mae"]
)


insurance_model3.fit(tf.expand_dims(x_train_normal,axis=-1),y_train,epochs=100)
```

```python
insurance_model3.evaluate(x_test_normal,y_test)
```
